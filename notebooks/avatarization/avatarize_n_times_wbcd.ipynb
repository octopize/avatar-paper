```python
from lsg import *

import pandas as pd
```

```python
wbcd = pd.read_csv("../../datasets/WBCD/breast_cancer_wisconsin.csv", sep=",")
wbcd = wbcd.astype("category")
wbcd = wbcd.drop(columns="Sample_code_number")

wbcd.head()
```

```python
res = []
for iter in range(25):

    wbcd_avatarize = avatarize(wbcd, k=20, ncp=2, drop_duplicates=False)
    wbcd_avatarize["iter"] = iter

    res.append(wbcd_avatarize)
```

```python
iteration = pd.concat(res, axis=0)
iteration.sample(10)
```

```python
iteration.to_csv("../../datasets/WBCD/wbcd_avatarized_25time_k20_nf_2.csv", index=False)
```
