```python
filename = os.listdir()
# filename.remove('.~lock.k801-150_avatarize.csv#')
df_list = []
for name in filename:
    df_temp = pd.read_csv(name)
    df_temp["id_ind"] = list(range(1, 684))
    df_temp["k"] = df_temp["k"]
    df_list.append(df_temp)
df_concat = pd.concat(df_list, axis=0)
```
