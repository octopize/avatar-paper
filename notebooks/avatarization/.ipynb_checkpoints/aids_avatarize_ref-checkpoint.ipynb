```python
from lsg import *

from lsg.dimension import *
from lsg.imputation import impute_missing
from lsg.evaluation_tools.visualization import projection_plot, distribution_plot, individual_travel_plot, correlation_plot, confidence_ellipse_plot
from lsg.evaluation_tools.tool_functions import get_stats
from lsg.security_metrics import *
from lsg.signal_metrics import *



import pandas as pd
import seaborn as sns
from sklearn.neighbors import NearestNeighbors
import numpy as np
import matplotlib.pyplot as plt


from math import exp

sns.set_theme(style="whitegrid")
import os
```

```python
aids = pd.read_csv("../../datasets/aids_original_data.csv", sep = ";")


categorical_val = []
continous_val = []
for column in aids.columns:
    if len(aids[column].unique()) <= 10:
        categorical_val.append(column)
    else:
        continous_val.append(column)

aids[categorical_val] = aids[categorical_val].astype('category')

aids = aids.drop(['pidnum'], axis=1)
```

## Avatarization

```python
aids_avatarize = avatarize(aids, k= 20, ncp= 5, drop_duplicates= False, distance_metric= 'minkowski')

aids_avatarize_imp = impute_missing(aids_avatarize)
```

# Security metrics 

```python
aids_avatarize_d = aids_avatarize.drop(columns= 'cd496', axis= 1)
aids_d = aids.drop(columns= 'cd496', axis= 1)
```
