```python
wbcd = wbcd.astype('category')
```

```python
wbcd_avatarize = avatarize(wbcd, k= 10, nf= 2, drop_duplicates= False )
```

```python
mt = Security_metrics()
mt.fit(wbcd, wbcd_avatarize)
print("local_cloaking (avatars_median): ", mt.local_cloaking['avatars_median'])
print("hidden_rate : ", mt.hidden_rate)

inference_dict = mt.inference_metrics(variable_list = ['Clump_Thickness', 'Uniformity_of_Cell_Size',
       'Uniformity_of_Cell_Shape', 'Bare_Nuclei', 'Bland_Chromatin'])
print( 'inference_hidden_rate : ', inference_dict['inference_hidden_rate'])
print( 'inference_local_cloaking : ', inference_dict['inference_local_cloaking'])
```
