```r
data <- read.csv(paste0("../datasets/aids_original_data.csv"), sep = ";", na.strings = c("NA", "", NA))
data <- data[,-1]
avatar <- read.csv(paste0("../datasets/aids_avatarized_base_k20_nf5.csv"), sep = ",", na.strings = c("NA", "", NA))
```

## Data and avatar imputation

```r
data_imp <- complete(mice(data,m=1,maxit=50,seed=500, printFlag = FALSE),1)
avatar_imp <- complete(mice(avatar,m=1,maxit=50,seed=500, printFlag = FALSE),1)
```

```r
data$cd496 <- data_imp$cd496
avatar$cd496 <- avatar_imp$cd496
```

## Figure 1a)

```r
data_tot <- rbind(data, avatar)
```

```r
pca <- PCA(data_tot, ncp = 5, graph = FALSE, ind.sup = (nrow(data_tot)/2 +1):nrow(data_tot))
res_ind <- as.data.frame(pca$ind$coord)
res_ind_sup <- as.data.frame(pca$ind.sup$coord)
```

```r
options(repr.plot.width = 8, repr.plot.height = 8)
if (save) {tiff(paste0(file_results, "HIV_pca_", Sys.Date(), ".tiff"), res = 600, width = 14, height = 14, units = "cm")}
scatter3D(res_ind[,1], res_ind[,2], res_ind[,3], pch = 21, 
          cex.axis = ifelse(save, 0.7, 1.2), cex.lab = ifelse(save, 0.8, 1.4), cex = ifelse(save, 0.6, 1.2), lwd = 0.5,
          col = "black", bg = "#696969", colkey = FALSE, 
          xlab = paste0("Axis 1 (", round(pca$eig[1,2], 2) ,"%)"), 
          ylab = paste0("Axis 2 (", round(pca$eig[2,2], 2) ,"%)"), 
          zlab = paste0("Axis 3 (", round(pca$eig[3,2], 2) ,"%)"), 
          ticktype = "detailed",
          phi = 20, # change l'angle de vue axe 3, par défaut 40
          theta = 35, # change l'angle de vue axe 1, par défaut 40
          bty = "f")
scatter3D(res_ind_sup[,1], res_ind_sup[,2], res_ind_sup[,3], add = TRUE, colkey = FALSE, 
          pch = 21, cex = ifelse(save, 0.6, 1.2), col = "black", bg = "#3BD6B0", lwd = 0.5)
legend("topleft", 
       legend = c("Patients", "Avatars"), pt.bg = c("#696969", "#3BD6B0"), 
       pch = 21, col = "black", pt.lwd = ifelse(save, 0.5, 1),
       title = NULL, cex = ifelse(save, 0.8, 1))
if (save) {dev.off()}
```

### Avatar

```r
summary_cox <- summary(coxph(Surv(time = days/7, event = cens)~factor(arms), data = avatar))
```

```r
val_arms0 <- "(Cox model reference)"
val_arms1 <- paste0("(HR [CI 95%]: ", format(round(as.vector(summary_cox$coefficients[1,2]), 2), nsmall=2), 
                    " [", format(round(as.vector(summary_cox$conf.int[1,3]), 2), nsmall=2), 
                    "-", format(round(as.vector(summary_cox$conf.int[1,4]), 2), nsmall=2), "])")
val_arms2 <- paste0("(HR [CI 95%]: ", format(round(as.vector(summary_cox$coefficients[2,2]), 2), nsmall=2), 
                    " [", format(round(as.vector(summary_cox$conf.int[2,3]), 2), nsmall=2), 
                    "-", format(round(as.vector(summary_cox$conf.int[2,4]), 2), nsmall=2), "])")
val_arms3 <- paste0("(HR [CI 95%]: ", format(round(as.vector(summary_cox$coefficients[3,2]), 2), nsmall=2), 
                    " [", format(round(as.vector(summary_cox$conf.int[3,3]), 2), nsmall=2), 
                    "-", format(round(as.vector(summary_cox$conf.int[3,4]), 2), nsmall=2), "])")
```

```r
options(repr.plot.width = 10.5, repr.plot.height = 8)
if (save) {tiff(paste0(file_results, "HIV_avatar_survival_table_", Sys.Date(), ".tiff"), width = 16, height = 12, units = "cm", res = 300)}
par(mfrow = c(1,1))
ggsurv <- ggsurvplot(survfit(Surv(time = days/7, event = cens)~factor(arms), data = avatar), 
                     data = avatar, 
                     color = "strata", palette = "grey", linetype = c(1,2,1,2),
                     risk.table = TRUE, risk.table.height = 0.25 ,
                     censor.shape="", censor.size = ifelse(save, 4, 5), size = ifelse(save, 0.5, 1), 
                     xlim = c(0,160), break.x.by = 20, 
                     tables.theme = theme_classic2(),
                     ggtheme = theme_classic(), 
                     risk.table.y.text = FALSE, risk.table.fontsize = ifelse(save, 3.5, 6),
                     xlab = "Time (week)", ylab = "Proportion not reaching\nprimary end point", 
                     legend = c(0.35,0.25), 
                     legend.title = "",
                     legend.labs = c(paste("Zidovudine", val_arms0), 
                                     paste("Zidovudine and didanosine", val_arms1), 
                                     paste("Zidovudine and zalcitabine", val_arms2), 
                                     paste("Didanosine", val_arms3)))


ggsurv$plot <- ggsurv$plot + 
  theme(legend.text = element_text(size = ifelse(save, 8, 15), color = "black"),
        axis.text = element_text(size = ifelse(save, 9, 15), color="black"),
        axis.title = element_text(size = ifelse(save, 10, 16), color = "black", face = "bold"),
        axis.title.y = element_text(vjust = 2))

ggsurv$table <- ggsurv$table + labs(
  title = "Number at risk") +
  theme(axis.title.x = element_blank(), 
        axis.text.x = element_blank(),
        axis.ticks.x = element_blank(),
        title = element_text(size = ifelse(save, 10, 14), color = "black", face = "bold"))

ggsurv
if (save) {dev.off()}
```

```r
options(repr.plot.width = 10.5, repr.plot.height = 6)
if (save) {tiff(paste0(file_results, "HIV_avatar_survival_", Sys.Date(), ".tiff"), width = 16, height = 10, units = "cm", res = 300)}
par(mfrow = c(1,1))
ggsurv <- ggsurvplot(survfit(Surv(time = days/7, event = cens)~factor(arms), data = avatar), 
                     data = avatar, 
                     color = "strata", palette = "grey", linetype = c(1,2,1,2),
                     risk.table = FALSE,
                     censor.shape="", censor.size = ifelse(save, 4, 5), size = ifelse(save, 0.5, 1), 
                     xlim = c(0,160), break.x.by = 20, 
                     ggtheme = theme_classic(), 
                     xlab = "Time (week)", ylab = "Proportion not reaching\nprimary end point", 
                     legend = c(0.35,0.2), 
                     legend.title = "",
                     legend.labs = c(paste("Zidovudine", val_arms0), 
                                     paste("Zidovudine and didanosine", val_arms1), 
                                     paste("Zidovudine and zalcitabine", val_arms2), 
                                     paste("Didanosine", val_arms3)))


ggsurv$plot <- ggsurv$plot + 
  theme(legend.text = element_text(size = ifelse(save, 7, 15), color = "black"),
        axis.text = element_text(size = ifelse(save, 9, 15), color="black"),
        axis.title = element_text(size = ifelse(save, 10, 16), color = "black", face = "bold"),
        axis.title.y = element_text(vjust = 2))

ggsurv
if (save) {dev.off()}
```
