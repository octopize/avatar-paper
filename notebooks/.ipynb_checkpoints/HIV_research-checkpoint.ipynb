```r
write_pvalue <- function(val, dec) {
  if(is.na(val)){
    return(NA)
  }else if (val < 0.001) {
    return("<0.001")
  } else {
    return(formatC(val, format = 'f', digits = dec))
  }
}
```

### Researchs on preanti distribution

```r
d <- density(data$preanti)
dydiff <- diff(d$y)
change <- c()
for (i in 1:(length(dydiff)-1)) {
    if (sign(dydiff[i]) != sign(dydiff[i+1])) {
        change <- c(change, i)
    }
}
```

```r
data["preantisep"] <- ifelse(data$preanti < d$x[change[2]], 0, 1)
```

```r
options(repr.plot.width = 24, repr.plot.height = 6)
par(mfrow = c(1,4))
plot(density(data$preanti), main = "Number of days of previously received antiretroviral therapyracerace")
plot(density(data$preanti))
lines(density(data[data$arms == 0, "preanti"]), col = "red")
lines(density(data[data$arms == 1, "preanti"]), col = "blue")
lines(density(data[data$arms == 2, "preanti"]), col = "green")
lines(density(data[data$arms == 3, "preanti"]), col = "orange")
plot(d)
for (i in change) {
    abline(v = d$x[change], col = "red")
}
plot(density(data$preanti))
lines(density(data[data$strat == 1, "preanti"]), col = "red")
lines(density(data[data$strat == 2, "preanti"]), col = "blue")
lines(density(data[data$strat == 3, "preanti"]), col = "green")
```

### Figure 1A

```r
diff = survdiff(Surv(time = days/7, event = cens)~factor(arms), data = avatar)
pval = pchisq(diff$chisq, length(diff$n)-1, lower.tail = FALSE)
pval = ifelse(pval < 0.001, "p < 0.001", paste("p =", round(pval,3)))
```

```r
options(repr.plot.width = 8, repr.plot.height = 8)
par(mfrow = c(1,1))
ggsurv <- ggsurvplot(survfit(Surv(time = days/7, event = cens)~factor(arms), data = avatar), 
                     data = avatar, 
                     risk.table = TRUE, pval = pval, risk.table.height = 0.25, pval.size = 5 ,
                     censor.shape="", censor.size = 4, size = 1.2, 
                     xlim = c(0,160), break.x.by = 20,
                     tables.theme = theme_classic2(),
                     ggtheme = theme_classic(), 
                     risk.table.y.text = FALSE, 
                     xlab = "Time (week)", ylab = "Survival probability (precised)", 
                     legend.title = "",
                     legend.labs = c("Zidovudine", "Zidovudine and\ndidanosine", "Zidovudine and\nzalcitabine", "Didanosine"))


ggsurv$plot <- ggsurv$plot + 
  theme(legend.text = element_text(size = 12, color = "black", face = "bold"),
        axis.text = element_text(size =12, color="black"),
        axis.title = element_text(size = 12, color = "black", face = "bold"),
        axis.title.y = element_text(vjust = 2))


ggsurv$plot <- ggpar(
  ggsurv$plot,
  legend = "top"
)

ggsurv$table <- ggsurv$table + labs(
  title = "Number at risk") +
  theme(axis.title.x = element_blank(), 
        axis.text.x = element_blank(),
        axis.ticks.x = element_blank(),
        title = element_text(size = 11, color = "black", face = "bold"))

ggsurv
```
